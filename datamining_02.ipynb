```python
import pandas as pd
```

```python
data=pd.read_csv('/content/Ship_Performance_Dataset.csv')
summary = data.describe()
print(summary)
```

```python
print(data.describe(include=['object']))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(data['Weekly_Voyage_Count'].dropna(),bins=20, edgecolor='black', color='skyblue')
plt.title('weekly voyage count distribution')
plt.xlabel('Weekly_Voyage_Count')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
data['Ship_Type'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=60)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(data['Weekly_Voyage_Count'].dropna(),patch_artist=True,boxprops=dict(facecolor='lightgreen'))

plt.title('BOX plot of weeklu voyage count')
plt.ylabel('Weekly_Voyage_Count')
plt.show()
```

```python
import seaborn as sns
```

```python
sns.pairplot(data[['Weekly_Voyage_Count','Speed_Over_Ground_knots','Distance_Traveled_nm','Draft_meters']].dropna())
plt.show()
```

```python
correlation_matrix=data[['Weekly_Voyage_Count','Speed_Over_Ground_knots','Distance_Traveled_nm','Draft_meters']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='Ship_Type', data=data, palette='Set2')
plt.title('ship type')
plt.xlabel('Ship_Type')
plt.ylabel('Count')
plt.show()
```

```python
sns.violinplot(x='Ship_Type', y='Weekly_Voyage_Count', data=data, palette='muted')
plt.title('Weekly Voyage Count by Ship Type')
plt.title('ship type by weekly voyage count')
plt.show()
```
